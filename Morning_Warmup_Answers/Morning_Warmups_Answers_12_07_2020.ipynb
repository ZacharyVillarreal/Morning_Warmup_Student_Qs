1. Write an SQL query to print details of the Workers whose FIRST_NAME ends with a.

**Answer**:<br>
    
SELECT * <br>
FROM Worker <br>
WHERE FIRST_NAME LIKE '%a'; <br>

2. Write an SQL query to print details of workers excluding first names, Vipul and Satish from Worker table.

**Answer**:<br>

SELECT * <br>
FROM Worker <br>
WHERE FIRST_NAME not in ('Vipul', 'Satish');<br>

3. Write an SQL query to fetch the number of workers for each department in the descending order.

**Answer**:<br>
    
SELECT DEPARTMENT, COUNT(WORKER_ID) AS No_of_workers<br>
FROM Worker <br>
GROUP BY DEPARTMENT <br>
ORDER BY No_of_workers DESC;<br>

4. Write an SQL query to fetch the departments that have less than five people in it.

**Answer**:<br>

Select DEPARTMENT, COUNT(WORKER_ID) AS 'Number of Workers' <br>
FROM Worker <br>
GROUP BY Department<br>
HAVING COUNT(WORKER_ID) < 5;<br>

5. Write an SQL query to print the first name (first_name) and last name (last_name) of the Workers and their position title (Worker_Title) who are also Managers.

**Answer**:<br>
    
SELECT FIRST_NAME, LAST_NAME, WORKER_TITLE <br>
FROM Worker AS w<br>
INNER JOIN Title AS t<br>
ON w.WORKER_ID = t.WORKER_REF_ID<br>
WHERE T.WORKER_TITLE LIKE 'Manager';<br>

6. Write an SQL query to show the second highest salary from a table.

**Answer**:<br>

SELECT MAX(Salary)<br>
FROM Worker<br>
WHERE Salary NOT IN <br>
    ( SELECT MAX(Salary) from Worker );
